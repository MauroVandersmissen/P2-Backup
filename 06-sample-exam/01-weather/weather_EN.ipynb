# Data analysis: Weather
In the file `weather_dataset.csv` you find weather data for various cities. It is up to you to clean and analyse this data. Use the provided code fields under the tasks to write down your code. 

Import the necessary packages.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

* Import the weather dataset into a DataFrame named `df_weather`
* Inspect the DataFrame by showing the first 10 rows (do not use slicing for this).


```python
filepath="weather_dataset.csv"
df_weather=pd.read_csv(filepath)
df_weather
```

* Change the dates in the DATE column to pandas datetime objects.
* Only include the rows with data from 2009


```python
df_weather["DATE"]=pd.to_datetime(df_weather["DATE"],format="%Y%m%d")
df_weather=df_weather[df_weather["DATE"].dt.year==2009]
df_weather
```

* Remove the rows with NaN values.

```python
df_weather.dropna()
df_weather
```

* Create 4 different DataFrames, one for each of the following cities: Basel, Oslo, Roma and Maastricht. Name them `df_Basel`, `df_Oslo`, `df_Roma` and `df_Maastricht` respectively.
* Each dataframe should include the columns with the dates and months and the columns with the data of the corresponding city. 



```python
# df_Basel=df_weather.loc["DATE":"BASEL_temp_max"]
# df_Oslo=df_weather.loc["OSLO_cloud_cover":"OSLO_temp_max"]
# df_Roma=df_weather.loc["ROMA_cloud_cover":"ROMA_temp_max"]
# df_Maastricht=df_weather.loc["MAASTRICHT_cloud_cover":"MAASTRICHT_max_temp"]
date=["DATE","MONTH"]
basel_columns=[col for col in df_weather.columns if "BASEL" in col]
basel_df=date+basel_columns
df_Basel=df_weather[basel_df]
print(df_Basel)
oslo_columns=[col for col in df_weather.columns if "OSLO" in col]
oslo_df=date+oslo_columns
df_Oslo=df_weather[oslo_df]
print(df_Oslo)
roma_columns=[col for col in df_weather.columns if "ROMA" in col]
roma_df=date+roma_columns
df_Roma=df_weather[roma_df]
print(df_Roma)
maas_columns=[col for col in df_weather.columns if "MAASTRICHT" in col]
maas_df=date+maas_columns
df_Maastricht=df_weather[maas_df]
print(df_Maastricht)
```

### Maastricht
Plot a figure displaying following information of the temperature measurements over time done in Maastricht:
* Temperature in °C is displayed on the y-axis
* Mean temperature in orange
* Max temperature in red
* Min temperature in blue
* Display a legend with appropriate labels, indicating which data is shown
* Ensure proper axis labels and a title are included
* On the x-axis, ensure there are indications for every month, labeled with the abbreviation of the month in English (Jan, Feb, March etc.) 

```python
plt.figure(figsize=(10,6))
plt.plot(df_Maastricht["DATE"],df_Maastricht["MAASTRICHT_temp_mean"],color="orange",label="Mean temp")
plt.plot(df_Maastricht["DATE"],df_Maastricht["MAASTRICHT_temp_max"],color="red",label="Max temp")
plt.plot(df_Maastricht["DATE"],df_Maastricht["MAASTRICHT_temp_min"],color="blue",label="Min temp")
plt.title("Temperature in °C in Maastricht over time")
plt.xlabel("Date")
plt.ylabel("Temperature in °C")
# plt.xticks(ticks=["2009-01","2009-02","2009-03","2009-04","2009-05","2009-06","2009-07","2009-08","2009-09","2009-10","2009-11","2009-12"],labels=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Okt","Nov","Dec"])
import matplotlib.dates as mdates
ax=plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b"))
plt.legend()
plt.show()
```

### Mean temperatures
Create a new graph and plot the mean temperatures over time of the cities for which you created seperate dataframes. Temperatures are in degrees Celcius and displayed on the y-axis.
* Maastricht in orange
* Roma in red
* Basel in green

* Display a legend with appropriate labels so it is clear what data is shown

```python
plt.figure(figsize = (10, 6))
plt.plot(df_Maastricht["DATE"],df_Maastricht["MAASTRICHT_temp_mean"],color="orange",label="Maastricht mean temp")
plt.plot(df_Roma["DATE"],df_Roma["ROMA_temp_mean"],color="red",label="Roma mean temp")
plt.plot(df_Basel["DATE"],df_Basel["BASEL_temp_mean"],color="green",label="Basel mean temp")
plt.plot(df_Oslo["DATE"],df_Oslo["OSLO_temp_mean"],color="blue",label="Oslo mean temp")
plt.title("Mean temperature in °C over time")
plt.xlabel("Date")
plt.ylabel("Temperature in °C")
import matplotlib.dates as mdates
ax=plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b"))
plt.legend()
plt.grid()
plt.show()
```
