## 1. Counting birds

### <u> Counting birds part 1 </u>

Every year there is a day where people are encouraged to count the birds in their garden. Joseph tries to participate every year. Below you see the data of Joseph's observations. Let's make a nice plot of the data, also include the total amount of birds spotted.

```python
# Sample data
data = {
    'year': ['2020', '2021', '2022', '2023', '2024'],
    'Pigeons': [33, 52, 26, 40, 38],
    'Sparrows': [54, 48, 66, 59, 61],
    'others': [104, 98, 88, 92, 101]
}
```

### Inspecting the data

Load and inspect the data.

```python
df=pd.DataFrame(data)
df
```

### Plotting the pigeon count

Plot the number of counted pigeons over the year, following these specifications:

* Line graph
* Size of the figure: (10,6)
* Full blue line, data point marked with a circle, linewidth 2, marker size 8
* Provide the necessary labels in order to interpret the graph and show a legend
* Provide a meaningfull title
* Show a grid

```python
fig,ax=plt.subplots(figsize=(10,6))
ax.plot(df["year"],df["Pigeons"],marker="o",color="blue",linestyle="-",linewidth=2,markersize=8,label="Pigeons")
ax.set_title("Bird Count")
ax.set_xlabel("Year")
ax.set_ylabel("Count")
ax.grid(True)

```

### Plotting the sparrow count

Plot the number of sparrows Joseph counted over the years, adhere to the specifications listed below:

### Bar chart of total
Display the data  using a bar chart (total) and a stacked bar chart (pigeons, sparrows, others). Ensure the total is displayed on the right side and the stacked bar on the left.
Make it informative (and pretty).

```python
bar_width=0.4
index=range(len(df["year"]))
fig,ax=plt.subplots(figsize=(10,6))
ax.bar(index,df["Pigeons"],bar_width,color="red",label="Pigeons")
ax.bar(index,df["Sparrows"],bar_width,bottom=df["Pigeons"],color="green",label="Sparrows")
ax.bar(index,df["others"],bar_width,bottom=df["Pigeons"]+df["Sparrows"],color="blue",label="others")
# ax.bar(index,df["total"],bar_width,color="purple",label="total")
ax.bar([i + bar_width for i in index], df['total'], bar_width, color='purple', label='total')
ax.set_title("Bird Count")
ax.set_xlabel("Year")
ax.set_ylabel("Count")
ax.set_xticks([i + bar_width / 2 for i in index], df['year'])
ax.legend()

# plt.figure(figsize=(10, 6))

# # Your code
# index = range(len(df['year']))
# bar_width = 0.35

# # Gestapelde bars
# plt.bar(index, df['Pigeons'], bar_width, color='red', label='Pigeons')
# plt.bar(index, df['Sparrows'], bar_width, bottom=df['Pigeons'], color='green', label='Sparrows')
# plt.bar(index, df['others'], bar_width, bottom=df['Pigeons'] + df['Sparrows'], color='blue', label='Others')

# # Voeg de "Birds counted" balk ernaast toe (verschuif index met bar_width)
# plt.bar([i + bar_width for i in index], df['total'], bar_width, color='purple', label='total')

# plt.xticks([i + bar_width / 2 for i in index], df['year'])

# plt.xlabel('Year')
# plt.ylabel('Count')

# plt.legend()
```

### <u> Counting birds part 4 </u>

### Grouped bar chart

Display the birds data using a grouped bar chart. Place the count of the pigeons and sparrows on the left of the x-tick, others and total on the right. Ensure the bars do not overlap.

```python
fix,ax=plt.subplots(figsize=(10,6))

bar_width = 0.2
x = range(len(df["year"]))
Pigeons = ax.bar([i - bar_width for i in x], df["Pigeons"], bar_width, color='red', label='Pigeons', align="center")
Sparrows = ax.bar(x, df["Sparrows"], bar_width, color='green', label='Sparrows', align="center")
others =  ax.bar([i + bar_width for i in x], df["others"], bar_width, color='blue', label='others', align="center")
total = ax.bar([i + 2*bar_width for i in x], df["total"], bar_width, color="purple", label="total", align="center")

#Let's add the values on top of the bars
ax.bar_label(Pigeons)
ax.bar_label(Sparrows)
ax.bar_label(others)
ax.bar_label(total)

ax.set_xticks(x, df["year"])
ax.set_title("Bird Count")

ax.set_xlabel("Year")
ax.set_ylabel("Count")
ax.legend()
ax
```
